```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset

athletes = pd.read_csv('C:/Users/HP WORLD/Desktop/PROJECTS/Olympic_history/athlete_events.csv');
regions = pd.read_csv('C:/Users/HP WORLD/Desktop/PROJECTS/Olympic_history/noc_regions.csv');
```

```python
athletes.head()
```

```python
regions.head()
```

```python
# Join the Dataframes

athletes_af = athletes.merge(regions,how = 'left', on = 'NOC')
athletes_af.head()
```

```python
athletes_af.shape
```

```python
# Inconsistent Column Names

athletes_af.rename(columns = {'region':'Region','notes':'Notes'}, inplace = True)
```

```python
# For Statistical Analysis

athletes_af.describe()
```

```python
# Check null Values

nan_values = athletes_af.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
# India Details

athletes_af.query('Team =="India"').head(5)
```

```python
# Norway Details

athletes_af.query('Team =="Norway"').head(5)
```

```python
# Top Countries Participating

top_10_countries = athletes_af.Team.value_counts().sort_values(ascending = False).head(10)
top_10_countries
```

```python
# Plot for the Top 10 Countries

plt.figure(figsize = (10,6))
plt.title('Overall Participation by Country')
sns.barplot(x = top_10_countries.index, y = top_10_countries, palette = 'Set1');
```

```python
# Age Distribution of Athletes

plt.figure(figsize = (10,6))
plt.title('Age Distribution of Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(athletes_af.Age, bins = np.arange(10,80,2), color = 'green', edgecolor = 'white');
```

```python
# Winter Olympic Sports

winter_sports = athletes_af[athletes_af.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
# Summer Olympic Sports

summer_sports = athletes_af[athletes_af.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
# Male and Female Participants

gender_counts = athletes_af.Sex.value_counts()
gender_counts
```

```python
# Pie Plot for Male and  Female Athletes

plt.figure(figsize =(9,5))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels = gender_counts.index, autopct = '%1.1f%%', startangle = 150, shadow = True);
```

```python
# Total Medals

athletes_af.Medal.value_counts()
```

```python
# Total Number of Female Athletes in each Olympic

female_participants = athletes_af[(athletes_af.Sex =='F') & (athletes_af.Season =='Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
# Gold Medal Athletes

goldMedal = athletes_af[(athletes_af.Medal =='Gold')]
goldMedal.head()
```

```python
# Gold Medals from each Country

goldMedal.Region.value_counts().reset_index(name ='Medal').head(10)
```

```python
totalGoldMedals = goldMedal.Region.value_counts().reset_index(name ='Medal').head(6)
g = sns.catplot(x='index', y='Medal', data = totalGoldMedals,
               height = 5, kind = 'bar',palette = 'rocket')
g.despine(left=True)
g.set_xlabels('Top 5 Countries')
g.set_ylabels('Number of Medals')
plt.title('Gold Medals per Country')
```

```python
not_null_medals = athletes_af[(athletes_af['Height'].notnull()) & (athletes_af['Weight'].notnull())]
```

```python
plt.figure(figsize = (12,10))
axis = sns.scatterplot(x = 'Height',y = 'Weight',data = not_null_medals, hue = 'Sex')
plt.title('Height Vs Weight of Olympic Medalists')
```
